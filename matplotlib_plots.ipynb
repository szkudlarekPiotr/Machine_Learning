### Plot showing function : y=f(x)=(a−4)x^2+(b−5)x+(c−6) with constants being numbers from my index (red), and function y=g(x)=e^x/e^x+1 (blue)

```python
import matplotlib.pyplot as plt
import numpy as np

# Część A:
a,b,c = 4,9,6
x = np.arange(-100.0, 100.0, 0.1)
y = ((a-4)*(x**2)) + ((b-5)*x) - (c-6)

fig = plt.figure()

zadanie = fig.add_subplot()

zadanie.set_xlabel('x')
zadanie.set_ylabel('y')
zadanie.axis([-20,20,-2,2])
zadanie.set_title('Wykres zadanie 2')

zadanie.plot(x, y, 'red')

# Część B:

y2 = np.exp(x)/(np.exp(x)+1)

zadanie.plot(x, y2, 'blue')
```

### 3D plot of function f(x,y)=−(x^2+y^3) on surface z=-(x^2+y^3)

```python
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
x = np.arange(-100,100, 5)
y = np.arange(-100,100, 5)
x,y = np.meshgrid(x,y)
z = -(x**2 + y**3)
surf = ax.plot_surface(x,y,z, cmap=cm.jet)
```
