```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('flats_for_clustering.tsv', sep="\t")

data = df.dropna()

data = data.replace(["parter", "niski parter"], 0)
data.loc[data["Pitro"] == "poddasze", "Pitro"] = data["Liczba piter w budynku"]+1

scaler = StandardScaler()
data_scaled = scaler.fit_transform(data)


kmeans = KMeans(n_clusters=5).fit(data_scaled)

pca = PCA(n_components=2)
dane_pca = pca.fit_transform(data_scaled)

plt.scatter(dane_pca[:,0], dane_pca[:, 1], c=kmeans.labels_)
plt.show()
```
